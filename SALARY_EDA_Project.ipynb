compare for key roles

Assign a variable to the grouped by jobtitle

```python
top_overtime = data.groupby('JobTitle')['OvertimePay'].mean().sort_values(ascending=False).head(10)
top_titles = top_overtime.index
print('Top overtime-heavy titles: \n', top_overtime)
```

FILTER AND CALCULATE AVERAGES

```python
data_top = data[data['JobTitle'].isin(top_titles)]
pay_comparison = data_top.groupby('JobTitle')[['BasePay', 'OvertimePay']].mean()
print('Average BasePay and Overtime: \n', pay_comparison)
```

```python
Plot results
```

```python
plt.style.use('fivethirtyeight')

fid, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(pay_comparison))

plt.bar(index, pay_comparison['BasePay'], bar_width, label='BasePay')
plt.bar([i + bar_width for i in index], pay_comparison['OvertimePay'], bar_width, label= 'Overtime Pay')

plt.xlabel('Job Title')
plt.ylabel('Average Pay ($)')
plt.title('Average Base Pay vs. Overtime Pay for Top Overtime-Heavy Job Titles')
plt.xticks([i + bar_width / 2 for i in index], pay_comparison.index, rotation=45, ha='right')
plt.legend()
plt.show()
```

This bar plot compares average BasePay and OvertimePay for 'Top Overtime-Heavy Job Titles'. Iterates through Job Titles, to plot two bars per job(BasePay and OvertimePay) side by side for comparison

Insights:

-BasePay Takes the Lead: As it significantly ranks higher compared to OvertimePay across all Jobs

-Overtime varies considerably, suggesting different overtime needs per practices of each job.

-Fire Department Extremes: Ass Chf of Dept(Fire Dept) and Battalion Chief, Fire Suppress show the highest BasePay but relatively low OvertimePay, indicating high base salaries but potentially less frequent overtime.

-Maintenance & Technical Jobs: Mech Shop & Equip Supt, Track Maint Wrk Sprv1, Transr Power Line Wrk Sprv1, Cable Maint Supervisor, exhibit more balanced Base/Overtime, SUGGESTING CONSISTENT NEED FOR OVERTIME WORK
